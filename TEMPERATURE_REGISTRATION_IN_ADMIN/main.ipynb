```python
#import necesary pkg
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read the file
df = pd.read_csv("FILES/datalog.csv", header=None, names=["Date", "Time", "Temperature (°C)", "Humidity (%)"])
print('STRUCTURE')
print(df.head(1))
```

```python
#Temperature plots
plt.figure(figsize=(10, 5))
plt.plot(df["Temperature (°C)"], label="Temperature", color="red")
plt.plot(df["Humidity (%)"], label="Humidity", color="blue")
plt.title("Measurement")
plt.xlabel("Data Index")
plt.ylabel("Values")
plt.grid()
plt.legend()
plt.show()
```
